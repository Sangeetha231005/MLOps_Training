**Machine Learning With Python : Save and Load the Trained Model**

```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```
df = pd.read_csv('/content/drive/MyDrive/Flood_detection/flood.csv')
```

```
df.head()
```

```
X = df[['MonsoonIntensity']]
y = df['FloodProbability']
```

```
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)
```

```
model = LinearRegression()
```

```
model.fit(X_train, y_train)
```

```
model.coef_
```

```
model.intercept_
```

```
prediction = model.predict(
    pd.DataFrame([[7]], columns=['MonsoonIntensity'])
)
print(prediction)
```

**Save Model To a File using Pickle :**

```
import pickle
```

```
with open('flood_model_pickle', 'wb') as file:
    pickle.dump(model, file)
```

```
with open('flood_model_pickle', 'rb') as file:
    mp = pickle.load(file)
```

```
mp.coef_
```

```
mp.intercept_
```

```
mp.predict(
    pd.DataFrame([[7]], columns=['MonsoonIntensity'])
)
print(prediction)
```

**Save Model using Joblib :**

```
import joblib
```

```
joblib.dump(model, 'flood_model_joblib')
```

```
mj = joblib.load('flood_model_joblib')
```
