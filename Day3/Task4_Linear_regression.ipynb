**Question** : Predict Canada's per capita income in year 2020.
There is an exercise folder on GitHub at the same level as this notebook. Download the canada_per_capita_income.csv file. Using this, build a regression model and predict the per capita income of Canadian citizens in year 2020.

```
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('/content/drive/MyDrive/Linear regression/Canada_per_capita.csv')
```

```
df.head()
```

```
%matplotlib inline

plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')

plt.scatter(df['year'], df['per capita income (US$)'],
            color='blue',
            marker='*')
```

```
model = LinearRegression()
```

```
model.fit(df[['year']], df['per capita income (US$)'])
```

```
model.predict(pd.DataFrame({'year':[2020]}))
```

```
model.coef_
```

```
model.intercept_
```

```
%matplotlib inline

plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')

plt.scatter(df['year'],
            df['per capita income (US$)'],
            color='red',
            marker='+')

plt.plot(df['year'],
         model.predict(df[['year']]),
         color='blue')
```
